## 1. Import Libraries and Load Data

```python
import numpy as np
import pandas as pd
# other than these two you should not import any other packages
```

## 1.1 Load and Display Data for 5_a.csv

```python
# write your code here
#printing the first 5 datapoints of 5_a.csv
A=pd.read_csv('')# add data
A.head()
```

# 2. Preprocess Data

## 2.1 Change 'proba' Feature Based on Threshold

```python
#changing the feature 'proba' as per instruction given above
A.loc[(A.proba)>=0.5,'proba']=1
A.loc[(A.proba)<0.5,'proba']=0
A.head()
```

# 3. Compute Performance Metrics

## 3.1 Compute Confusion Matrix

```python
#computing Confusion Matrix
TN_df=A[(A['y']==0)&(A['proba']==0)]
TN=len(TN_df)
TP_df=A[(A['y']==1)&(A['proba']==1)]
TP=len(TP_df)
FN_df=A[(A['y']==1)&(A['proba']==0)]
FN=len(FN_df)
FP_df=A[(A['y']==0)&(A['proba']==1)]
FP=len(FP_df)

print("Confusion Matrix : ",[[TN,FP],[FN,TP]])

#computing F1 SCore
P=FN+TP
N=TN+FP
con_mat=[[TN,FN],[FP,TP]]
Precision=TP/(TP+FP)
Recall=TP/P
F1_score=2*((Precision*Recall)/(Precision+Recall))
print("F1 SCORE : ",F1_score)

#computing Accuracy Score
Acc_score=(TP+TN)/(N+P)
print("Accuracy Score : ",Acc_score)

#import matplotlib.pyplot as plt
#%matplotlib inline

#Readind the original .csv file again
A2=pd.read_csv('')#read .csv file
Uniq_values=A2.proba.unique() #Unique values of y^
Uniq_values=sorted(Uniq_values,reverse=True) #sorting y^

#computing AUC Score
TPR_arr=[]
FPR_arr=[]
for i in Uniq_values:
    P=0
    N=0
    TPR=FPR=0
    A2.loc[A2['proba']>i,'y_hat']=1
    A2.loc[A2['proba']<=i,'y_hat']=0
    TN_df=A2[(A2['y']==0)&(A2['y_hat']==0)]
    TN=len(TN_df)
    TP_df=A2[(A2['y']==1)&(A2['y_hat']==1)]
    TP=len(TP_df)
    FN_df=A2[(A2['y']==1)&(A2['y_hat']==0)]
    FN=len(FN_df)
    FP_df=A2[(A2['y']==0)&(A2['y_hat']==1)]
    FP=len(FP_df)
    P=FN+TP
    N=TN+FP
    TPR=TP/P
    FPR=FP/N
    TPR_arr.append(TPR)
    FPR_arr.append(FPR)

TPR_arr=np.array(TPR_arr)
FPR_arr=np.array(FPR_arr)

#plt.plot(TPR_arr,FPR_arr)
AUC=np.trapz(TPR_arr,FPR_arr)
print("AUC SCORE :  ",AUC)
```
